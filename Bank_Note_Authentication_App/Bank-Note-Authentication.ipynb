## Bank Note Authentication Problem

Find the Dataset here `https://www.kaggle.com/ritesaluja/bank-note-authentication-uci-data`

```python
# Importing the libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle

import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the dataset into a dataframe
data = pd.read_csv('BankNote_Authentication.csv')
```

```python
# Checking the data
data.head()
```

```python
# Describing the data
data.describe(percentiles=[0.25,0.5,0.75])
```

```python
# Checking the null values
data.isnull().sum()
```

```python
# Divide the dependent and independent variables
X = data.iloc[:,:-1]
y = data.iloc[:,-1:]
```

```python
X.head(20)
```

```python
y.head(20)
```

```python
# Creating the train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Implementing RandomForestClassifier
classifier = RandomForestClassifier()

classifier.fit(X_train, y_train)
```

```python
# Predict the test data
y_pred = classifier.predict(X_test)
```

```python
# Check the accuracy score of the prediction 
score = accuracy_score(y_pred, y_test)
```

```python
score
```

```python
# Convert the model into pickle files
pickle_out = open("predictor.pkl", 'wb')
pickle.dump(classifier, pickle_out)
pickle_out.close()
```
